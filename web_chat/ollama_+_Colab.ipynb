```python
from ollama import chat
```

```python
def ask_llm(prompt, history=None,model='gemma3'):
    print('calling LLM')
    messages = []
    if history:
        for u, b in history:
            messages.append({"role": "user", "content": u})
            messages.append({"role": "assistant", "content": b})
    messages.append({"role": "user", "content": prompt})

    response = chat(
        model=model,
        messages=messages,
        stream=False
    )
    print("done colling")
    return response["message"]["content"]
```

```python
ask_llm("who is pikachu? maximum 20 words")
```

```python
def start_chat(model='gemma3'):
  history = []
  while True:
    prompt = input("\nask-> ")
    if prompt in ('/bye', 'exit'): break
    response = ask_llm(prompt=prompt,history=history,model=model)
    print("\n",response)
    history.append((prompt,response))
  return history
```

```python
app = Flask(__name__)
```

```python
history = []
```

```python
@app.route("/", methods=["GET", "POST"])
def chat_llm():
    global history
    model = "gemma3"

    if request.method == "POST":
        prompt = request.form["prompt"]

        if prompt not in ("/bye", "exit"):
            response = ask_llm(prompt, history, model)
            print(history)
            history.append((prompt, response))

    return render_template_string(HTML, history=history)
```

```python
# Run server
if __name__ == "__main__":
    app.run(host="0.0.0.0", port=1211,debug=False)
```
